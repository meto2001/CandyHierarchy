## Q1:

```python
data['Q1: GOING OUT?'].unique()
```

```python
Q1_mode = data["Q1: GOING OUT?"].mode()[0]
data['Q1: GOING OUT?'].fillna(Q1_mode, inplace = True)
data['Q1: GOING OUT?'] = data['Q1: GOING OUT?'].astype("category")
```

## Q2:

```python
Q2_mode = data["Q2: GENDER"].mode()[0]
data['Q2: GENDER'].fillna(Q2_mode, inplace = True)
data['Q2: GENDER'] = data['Q2: GENDER'].astype("category")
data['Q2: GENDER'].value_counts()
```

## Q3:

```python
data['Q3: AGE'].unique()
```

```python
data['Q3: AGE'].replace(['Old enough','45-55','24-50','?','no','Many','hahahahaha','older than dirt','Enough','See question 2','old','ancient','old enough','5u','46 Halloweens.','sixty-nine','Over 50','OLD','MY NAME JEFF','59 on the day after Halloween','your mom','I can remember when Java was a cool new language', '60+', '1000', '70 1/2', '1', '4'],np.nan,inplace=True)
Q3_mean = data['Q3: AGE'].mean()
data['Q3: AGE'].fillna(Q3_mean, inplace = True)
data['Q3: AGE'] = data['Q3: AGE'].astype("int")
```

## Q4:

```python
data["Q4: COUNTRY"] = data["Q4: COUNTRY"].str.strip(".")
data["Q4: COUNTRY"] = data["Q4: COUNTRY"].str.strip()
data["Q4: COUNTRY"] = data["Q4: COUNTRY"].str.upper()
```

```python
countries = ['USA','CANADA','UK','UNITED STATES']
for country in countries:
    matches = process.extract(country, data['Q4: COUNTRY'], limit = data.shape[0])
    for match in matches:
        if (match[1] >= 70):
            data.loc[data['Q4: COUNTRY'] == match[0],'Q4: COUNTRY'] = country
```

```python
data['Q4: COUNTRY'].replace(['USA','AMERICA','US OF A','I PRETEND TO BE FROM CANADA, BUT I AM REALLY FROM THE UNITED STATES','USSA','USA? HARD TO TELL ANYMORE','U S','U.S.A','U S A',"'MERICA",'AHEM....AMERCA','USA USA USA!!!!','U.S.','USAS','N. AMERICA','NEW YORK','UNITED STATSS','UNITE STATES','NEW JERSEY','NORTH CAROLINA',"I DON'T KNOW ANYMORE","INSANITY LATELY","CALIFORNIA","35","1","32","45","46","UD","EUROPE","FEAR AND LOATHING","NARNIA","SOVIET CANUCKISTAN","SUBSCRIBE TO DM4UZ3 ON YOUTUBE","EARTH"],'UNITED STATES',inplace=True)
data['Q4: COUNTRY'].replace(["UK","ENGLAND","ENDLAND","U.K."],'UNITED KINGDOM',inplace=True)
data['Q4: COUNTRY'].fillna("UNITED STATES", inplace = True)
data['Q4: COUNTRY'] = data['Q4: COUNTRY'].astype("category")
data['Q4: COUNTRY'].value_counts()
```

## Q5:

```python
data["Q5: STATE, PROVINCE, COUNTY, ETC"].value_counts()
```

```python
data["Q5: STATE, PROVINCE, COUNTY, ETC"] = data["Q5: STATE, PROVINCE, COUNTY, ETC"].str.strip(".")
data["Q5: STATE, PROVINCE, COUNTY, ETC"] = data["Q5: STATE, PROVINCE, COUNTY, ETC"].str.strip()
data["Q5: STATE, PROVINCE, COUNTY, ETC"] = data["Q5: STATE, PROVINCE, COUNTY, ETC"].str.upper()
```

```python
matches = process.extract('NEW YORK',data["Q5: STATE, PROVINCE, COUNTY, ETC"], limit = len(data["Q5: STATE, PROVINCE, COUNTY, ETC"]))
for match in matches:
    if match[1]>=80:
        data.loc[data["Q5: STATE, PROVINCE, COUNTY, ETC"] == match[0], "Q5: STATE, PROVINCE, COUNTY, ETC"] = 'NEW YORK'
data["Q5: STATE, PROVINCE, COUNTY, ETC"].replace(["NY","NY","NYC"],'NEW YORK',inplace=True)
```

```python
matches = process.extract("CALIFORNIA",data["Q5: STATE, PROVINCE, COUNTY, ETC"], limit = len(data["Q5: STATE, PROVINCE, COUNTY, ETC"]))
for match in matches:
    if match[1]>=80:
        data.loc[data["Q5: STATE, PROVINCE, COUNTY, ETC"] == match[0], "Q5: STATE, PROVINCE, COUNTY, ETC"] = 'CALIFORNIA'
```

```python
data["Q5: STATE, PROVINCE, COUNTY, ETC"].replace(["1","48","see question2","ur mom","Nope","LONDOM"],np.nan ,inplace = True)
data["Q5: STATE, PROVINCE, COUNTY, ETC"].fillna("CALIFORNIA", inplace = True)
data["Q5: STATE, PROVINCE, COUNTY, ETC"] = data["Q5: STATE, PROVINCE, COUNTY, ETC"].astype("string")
data["Q5: STATE, PROVINCE, COUNTY, ETC"].unique()
```
